```python
import matplotlib.pyplot as plt

# Define age ranges and their respective percentages
age_ranges = ['35-44', '45-54', '55-64', '65+']
percentages = [10, 27.2, 42.1, 20.8]

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(percentages, labels=age_ranges, autopct='%1.1f%%', startangle=140)
plt.title('Age Distribution of YouTube Viewers')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Define age ranges and their respective percentages
age_ranges = ['35-44', '45-54', '55-64', '65+']
percentages = [10, 27.2, 42.1, 20.8]

# Plot the bar chart
plt.figure(figsize=(8, 6))
plt.bar(age_ranges, percentages, color='skyblue')
plt.title('Percentage of YouTube Viewers by Age Range')
plt.xlabel('Age Range')
plt.ylabel('Percentage')
plt.ylim(0, 50)  # Adjust ylim based on the maximum percentage
plt.grid(axis='y')
plt.show()
```
